```python
import psycopg2 as pg
import pandas as pd

import os
config = os.environ

engine = pg.connect(
    "user='{0}' password='{1}' host='{2}' port='{3}'".format(
        config['DB_USER'],
        config['DB_PASSWORD'],
        config['DB_HOST'],
        config['DB_PORT']
    )
)

df = pd.read_sql('select * from SPY', con=engine)
```

```python
x = df.loc[:,"timestamp"]
y = df.loc[:,"LAST_PRICE"]
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
